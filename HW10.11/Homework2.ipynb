Парадигма мешка слов - это один  из самых эффективных методов парсинга текстовых данных. На занятии мы получили следующий мешок:

```python
try:
    fhand = open('text.txt')
except:
    print('Error in :', 'text.txt')
    exit()
    
counts = {}
for line in fhand:
    words = line.lower().split()
    for i in words:
        counts[i] = counts.get(i,0) +1
print(counts)
```
