## Об аргументах

Они бывают следующие:

1)Required arguments (важен порядок аргументов)

2) Keyword arguments (важен тип аргументов)

3) Default arguments (по умолчанию)

4) Variable-length arguments (по длине аргумента)

```python
#Example 2 (Required arguments)

def  rec_func(a,b,c):
    return a+b/c

print(rec_func(10,20,30))
```

```python
#Example 3 (Keyword arguments)

def print_str(str):
    print(str + ',Yo')
    
print_str('Hey Hey')
print_str(10)
```

```python
#Example 4(Default arguments)

def printInfo(name, age = 35):
    print('Name is :', name)
    print('Age is :', age)
    
    
printInfo('Jack')
```

```python
#Example 5 (Variable length arguments)

def new_info(arg1, *vartuple):
    print(arg1)
    for i in vartuple:
        print(i)
        
#new_info(10)

#print(new_info(10,20,30))

new_info(10,20,30)
```
