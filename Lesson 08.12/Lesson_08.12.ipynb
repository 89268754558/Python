# Объектно-ориентированное программирование (ООП) 

**ООП**— парадигма программирования, в которой основными концепциями являются понятия **объектов** и **классов**.

**Класс** — тип, описывающий устройство объектов. 

**Объект** — это экземпляр класса. 


Класс можно сравнить с чертежом, по которому создаются объекты.

```python
#First simple class


class SimpleClass:
    pass


a = SimpleClass()   # Экземпляр 1
b = SimpleClass()   # Экземпляр 2
```

```python
a.arg = 1   #Добавим ЭКЗЕМПЛЯРУ 1 атрибут .arg
b.arg = 2  #Добавим ЭКЗЕМПЛЯРУ 2 атрибут .arg
```

```python
print(a.arg)
print(b.arg)
```

```python
c = SimpleClass()
print(c.arg)
```

## У класса можно задавать собственные методы

```python
class MediumClass:
    def my_func(self):
        return 'Hello World'
```

```python
a = MediumClass()
a.my_func() # self - передает экземпляр класса MediumClass
```

**Инкапсуляция** — ограничение доступа к составляющим объект компонентам (методам и переменным).


Инкапсуляция делает некоторые из компонент доступными только внутри класса.

```python
class A:
    def _private(self):    # одиночное подчеркивание показывает приватность 
        print("This is private!")
        
        
```

```python
a1 = A()
a1._private()
```

```python
class B:
    def __private(self):  ## двойное подчеркивание делает его недоступным
        return ("This is private!")
```

```python
b1 = B()
b1.__private()
```

## Но есть лазейка 

```python
b1._B__private()
```

# Наследование

**Наследование** подразумевает то, что дочерний класс содержит все атрибуты родительского класса, при этом некоторые из них могут быть переопределены или добавлены в дочернем. 



```python
#Вспомним словари

my_dict = {}
my_dict['kek'] = my_dict.get('kek') 
```

```python
my_dict

## Следовательно .get ставит по умолчанию None
```

```python
class Mydict(dict):
    def get(self, key, default = 0):
        return dict.get(self, key, default)
first_dict = dict()
second_dict = Mydict()
```

```python
first_dict['Kek'] = 0
second_dict['Kek'] = 0
```

```python
first_dict
```

```python
second_dict
```

```python
first_dict['Lol'] = first_dict.get('Lol')
second_dict['Lol'] = second_dict.get('Lol')
```

```python
second_dict 
```

```python
first_dict
```

# Полиморфизм

**Полиморфизм** - разное поведение одного и того же метода в разных классах.

Например, мы можем сложить два числа, и можем сложить две строки.

При этом получим разный результат, так как числа и строки являются разными классами.

```python
1 + 1
```

```python
'1' + '1' 
```
